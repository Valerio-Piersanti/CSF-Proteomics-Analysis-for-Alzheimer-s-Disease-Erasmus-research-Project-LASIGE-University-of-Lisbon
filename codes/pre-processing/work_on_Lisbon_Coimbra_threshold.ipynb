Refining Column Headers for Enhanced Sample Identification

Adjusting column metadata to improve the readability of patient identifiers (IDs) and ensure consistent data visualization.

```python
import re

def extract_id(col):
    # caso LIS
    m1 = re.search(r"LIS\d+", col)
    if m1:
        return m1.group()
    
    # caso Coimbra numerico
    m2 = re.search(r'_(\d+)_', col)
    if m2:
        return m2.group(1)
    
    return col

df.columns = [extract_id(c) for c in df.columns]
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 1. Preparazione dati
X = data_log2.drop(columns=["Lab"]).values
labels_lab = data_log2["Lab"].values

# Mappatura dei nomi per la legenda
# Assumo: 0 -> Coimbra (Salmon), 1 -> Lisbon (Skyblue) per coerenza con il boxplot
label_map = {0: 'Coimbra', 1: 'Lisbon'}
color_map = {0: 'red', 1: 'blue'}

# 2. Z-score e PCA
X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
pc = pca.fit_transform(X_scaled)

# 3. Plot
plt.figure(figsize=(8,6))

for lab_val in sorted(set(labels_lab)):
    idx = labels_lab == lab_val
    plt.scatter(
        pc[idx, 0], 
        pc[idx, 1], 
        label=label_map[lab_val], # Usa il nome della città
        color=color_map[lab_val], # Usa lo stesso colore del boxplot
        alpha=0.8,
        edgecolors='w' # Aggiunge un bordino bianco per far risaltare i punti
    )

plt.legend(title="City")
plt.title("PCA – Protein Expression Analysis", fontsize=14)
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)")

plt.grid(True, linestyle='--', alpha=0.5) # Un po' di griglia aiuta a leggere le distanze
plt.savefig('PCA_City_Colors.png')
plt.show()
```

```python
df_ad_ct = df_all_data[(df_all_data['Groups']=='MCI-AD') | (df_all_data['Groups']=='MCI-CT')]
```

```python
ad = df_ad_ct[df_ad_ct['Groups']=='MCI-AD']
ad_prot = ad.drop(columns=['Groups'])
ad_prot = ad_prot.drop(columns=['Gender'])
ad_prot = ad_prot.drop(columns=['Age'])
ad_prot = ad_prot.drop(columns=['Sample code'])
ad_prot = ad_prot.drop(columns=['Lab'])
```

```python
ct = df_ad_ct[df_ad_ct['Groups']=='MCI-CT']
ct_prot = ct.drop(columns=['Groups'])
ct_prot = ct_prot.drop(columns=['Gender'])
ct_prot = ct_prot.drop(columns=['Age'])
ct_prot = ct_prot.drop(columns=['Sample code'])
ct_prot = ct_prot.drop(columns=['Lab'])
```

```python
log2fc = ad_prot.mean() - ct_prot.mean()
```

```python
from scipy.stats import ttest_ind
p_values = []

for protein in ad_prot.columns:
    stat, p = ttest_ind(ad_prot[protein], ct_prot[protein], nan_policy="omit")
    p_values.append(p)

p_values = pd.Series(p_values, index=ad_prot.columns)
```

```python
results = pd.DataFrame({
    "log2FC": log2fc,
    "pvalue": p_values
})

results["neg_log10_p"] = -np.log10(results["pvalue"])
import numpy as np

results["color"] = "grey"

results.loc[(results["pvalue"] < 0.05) & (results["log2FC"] > 0), "color"] = "red"
results.loc[(results["pvalue"] < 0.05) & (results["log2FC"] < 0), "color"] = "blue"
```

```python
plt.figure(figsize=(10,8))
proteins_to_label = [
    "P05546","Q14624","P02766","P13473","P02747",
    "P14618","P07333","P02647","P25311","P27169"
]
# non significative
plt.scatter(
    results.loc[results["color"] == "grey", "log2FC"],
    results.loc[results["color"] == "grey", "neg_log10_p"],
    color="grey",
    alpha=0.5,
    label="Not significant"
)

# up in AD
plt.scatter(
    results.loc[results["color"] == "red", "log2FC"],
    results.loc[results["color"] == "red", "neg_log10_p"],
    color="red",
    alpha=0.8,
    label="Up in AD"
)

# up in CT
plt.scatter(
    results.loc[results["color"] == "blue", "log2FC"],
    results.loc[results["color"] == "blue", "neg_log10_p"],
    color="blue",
    alpha=0.8,
    label="Up in CT"
)

plt.axhline(-np.log10(0.05), linestyle="--")
plt.axvline(0, linestyle="--")

plt.xlabel("log2 Fold Change (AD vs CT)")
plt.ylabel("-log10(p-value)")
plt.title("Volcano Plot MCI-AD vs MCI-CT Lisbon+Coimbra PLS-DA")
plt.legend()
plt.savefig('Volcano_Plot_MCI-AD_vs_MCI-CT_LC.png')
for protein in proteins_to_label:
    if protein in results.index:
        x = results.loc[protein, "log2FC"]
        y = results.loc[protein, "neg_log10_p"]

        plt.text(
            x, y,
            protein,
            fontsize=9,
            ha='right',
            bbox=dict(facecolor='white', alpha=0.6, edgecolor='none')
        )
plt.savefig('Volcano_Plot_plsda_LC.png')
plt.show()
```

Data Normalization
The dataset was normalized using Z-score to bring all protein levels onto a comparable scale.

Process: Each value was adjusted to center the data around a mean of 0 with a standard deviation of 1.

Result: This ensures that proteins with higher expression levels do not unfairly dominate the analysis.

```python
data_log2_norm = data_log2.copy()
data_log2_norm = data_log2_norm[data_log2_norm.columns.drop('Lab')]
z_score_df = (data_log2_norm-data_log2_norm.mean())/data_log2_norm.std()
```
