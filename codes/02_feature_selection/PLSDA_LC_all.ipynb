```python
def calculate_vip(pls, X):
    
    t = pls.x_scores_
    w = pls.x_weights_
    q = pls.y_loadings_
    
    p, h = w.shape
    s = np.diag(t.T @ t @ q.T @ q).reshape(h, -1)
    
    Wnorm = (w ** 2) / np.sum(w ** 2, axis=0)
    
    vip = np.sqrt(p * (Wnorm @ s) / np.sum(s))
    
    return vip.flatten()
```

Dataset: Coimbra + Lisbon (Merged)
In this section, a Partial Least Squares Discriminant Analysis (PLS-DA) is performed on the combined clinical proteomic dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, matthews_corrcoef
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.cross_decomposition import PLSRegression
from sklearn.preprocessing import LabelEncoder

# ======================
# DATA LOADING
# ======================
X = data_log2
y = np.array(list_groups)

# ======================
# GLOBAL PARAMETERS
# ======================
vip_thresholds = [0.1, 0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
decision_thresholds = [0.4, 0.45, 0.5, 0.55, 0.6]

n_iterations = 50  
n_seeds = 15       

all_test_mcc = []
all_test_auc = []
all_n_features = []
all_best_vip_thr = []
all_best_dec_thr = []
global_stable_counts = pd.Series(0, index=X.columns)

def stratified_bootstrap(X, y):
    ad_idx = np.where(y == "MCI-AD")[0]
    ct_idx = np.where(y == "MCI-CT")[0]
    ad_sample = np.random.choice(ad_idx, size=len(ad_idx), replace=True)
    ct_sample = np.random.choice(ct_idx, size=len(ct_idx), replace=True)
    indices = np.concatenate([ad_sample, ct_sample])
    np.random.shuffle(indices)
    return X.iloc[indices], y[indices]

# ======================
# MAIN SEED LOOP
# ======================
for seed in range(n_seeds):
    print(f"\n" + "="*40)
    print(f"RUNNING SEED {seed}")
    print("="*40)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, stratify=y, random_state=seed
    )

    cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=seed)
    le = LabelEncoder()
    le.fit(y_train)
    y_train_enc = le.transform(y_train)
    y_test_enc = le.transform(y_test)

    # Griglia per ottimizzazione incrociata (VIP x Decision)
    grid_mcc = np.zeros((len(vip_thresholds), len(decision_thresholds)))
    grid_auc = np.zeros(len(vip_thresholds))

    for v_idx, v_thr in enumerate(vip_thresholds):
        fold_mccs = []
        fold_aucs = []

        for train_idx, val_idx in cv.split(X_train, y_train):
            X_sub, y_sub = X_train.iloc[train_idx], y_train[train_idx]
            X_val, y_val = X_train.iloc[val_idx], y_train[val_idx]
            y_val_enc = le.transform(y_val)

            # --- Feature Selection (VIP Stability) ---
            vip_matrix = np.zeros((n_iterations, X_sub.shape[1]))
            for i in range(n_iterations):
                X_b, y_b = stratified_bootstrap(X_sub, y_sub)
                pls_b = PLSRegression(n_components=2)
                pls_b.fit(X_b, le.transform(y_b))
                vip_matrix[i] = (calculate_vip(pls_b, X_b) > 1).astype(int)

            vip_freq = vip_matrix.mean(axis=0)
            sel_feats = X_sub.columns[vip_freq >= v_thr]
            if len(sel_feats) == 0: sel_feats = [X_sub.columns[np.argmax(vip_freq)]]

            # --- Validation ---
            pls_cv = PLSRegression(n_components=2)
            pls_cv.fit(X_sub[sel_feats], le.transform(y_sub))
            y_prob_val = pls_cv.predict(X_val[sel_feats]).ravel()

            mccs_per_dec = [matthews_corrcoef(y_val_enc, (y_prob_val > d_thr).astype(int)) for d_thr in decision_thresholds]
            fold_mccs.append(mccs_per_dec)
            fold_aucs.append(roc_auc_score(y_val_enc, y_prob_val))

        grid_mcc[v_idx, :] = np.mean(fold_mccs, axis=0)
        grid_auc[v_idx] = np.mean(fold_aucs)

    # Identificazione best combination
    best_v_idx, best_d_idx = np.unravel_index(np.argmax(grid_mcc), grid_mcc.shape)
    best_v_thr = vip_thresholds[best_v_idx]
    best_d_thr = decision_thresholds[best_d_idx]

    # --- Training Finale (Full X_train) ---
    final_vip_matrix = np.zeros((n_iterations, X_train.shape[1]))
    for i in range(n_iterations):
        X_b, y_b = stratified_bootstrap(X_train, y_train)
        pls_b = PLSRegression(n_components=2).fit(X_b, le.transform(y_b))
        final_vip_matrix[i] = (calculate_vip(pls_b, X_b) > 1).astype(int)
    
    final_vip_freq = final_vip_matrix.mean(axis=0)
    top_features = X_train.columns[final_vip_freq >= best_v_thr]
    if len(top_features) == 0: top_features = [X_train.columns[np.argmax(final_vip_freq)]]

    global_stable_counts.loc[top_features] += 1

    # --- Test Finale (X_test) ---
    pls_final = PLSRegression(n_components=2).fit(X_train[top_features], y_train_enc)
    y_prob_test = pls_final.predict(X_test[top_features]).ravel()
    
    test_mcc = matthews_corrcoef(y_test_enc, (y_prob_test > best_d_thr).astype(int))
    test_auc = roc_auc_score(y_test_enc, y_prob_test)

    all_test_mcc.append(test_mcc)
    all_test_auc.append(test_auc)
    all_n_features.append(len(top_features))
    all_best_vip_thr.append(best_v_thr)
    all_best_dec_thr.append(best_d_thr)

    print(f"Seed {seed}: MCC={test_mcc:.3f}, AUC={test_auc:.3f}, Features={len(top_features)}")

# ======================
# FINAL PERFORMANCE SUMMARY
# ======================
print("\n" + "="*50)
print("FINAL CONSOLIDATED RESULTS (30 SEEDS)")
print("="*50)
print(f"Mean MCC: {np.mean(all_test_mcc):.4f} (±{np.std(all_test_mcc):.4f})")
print(f"Mean AUC: {np.mean(all_test_auc):.4f} (±{np.std(all_test_auc):.4f})")
print(f"Mean Features: {np.mean(all_n_features):.2f} (±{np.std(all_n_features):.2f})")

print("\n--- TOP 30 MOST STABLE PROTEINS ACROSS SEEDS ---")
global_stability_LC = (global_stable_counts / n_seeds).sort_values(ascending=False)
print(global_stability_LC)

print("\n--- FREQUENCY OF SELECTED VIP THRESHOLDS (Stability) ---")
print(pd.Series(all_best_vip_thr).value_counts(normalize=True).sort_index())

df_plsda_LC = pd.DataFrame({
    "model": "PLS-DA",
    "seed": list(range(n_seeds)),
    "mcc": all_test_mcc,
    "auc": all_test_auc,
    "n_features": all_n_features,
    "best_vip_threshold": all_best_vip_thr,
    "best_decision_threshold": all_best_dec_thr
})

print("\n--- FREQUENCY OF SELECTED DECISION THRESHOLDS (Classification) ---")
print(pd.Series(all_best_dec_thr).value_counts(normalize=True).sort_index())
import seaborn as sns

import seaborn as sns
import matplotlib.pyplot as plt

# Impostiamo uno stile pulito per tutti i grafici
sns.set_theme(style="whitegrid")
```

Training: Coimbra Dataset | Testing: Lisbon Dataset
This section describes the application of Partial Least Squares Discriminant Analysis (PLS-DA) to evaluate the generalizability of the protein signature across different clinical centers (multicentric validation).

```python
# ======================
# DATA LOADING
# ======================
df_all_data = pd.read_pickle('df_all_data_Lisbon_Coimbra_threshold.pkl')

df_Coimbra = df_all_data[df_all_data['Lab_name'] == 'Coimbra']
df_Lisbon = df_all_data[df_all_data['Lab_name'] == 'Lisbon']

cols_to_drop = ['Lab', 'Sample code', 'Groups', 'Age', 'Gender', 'Lab_name']

# --- DEFINIZIONE TRAIN E TEST ---
X_train = df_Coimbra.drop(columns=cols_to_drop)
y_train = df_Coimbra['Groups'].values

X_test = df_Lisbon.drop(columns=cols_to_drop)
y_test = df_Lisbon['Groups'].values

# ======================
# PARAMETERS
# ======================
thresholds = [0.1, 0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]
decision_thresholds = [0.40, 0.45, 0.50, 0.55, 0.60]
n_iterations = 50
n_seeds = 15
n_iterationsf = 50 

all_test_mcc = []
all_test_auc = []
best_stability_thresholds = [] # VIP threshold
best_decision_thresholds = []  # Classifier threshold
n_features_per_seed = [] 

# Nuova variabile richiesta
global_stable_counts_coimbraonlisbon = pd.Series(0, index=X_train.columns)

def stratified_bootstrap(X, y):
    ad_idx = np.where(y == "MCI-AD")[0]
    ct_idx = np.where(y == "MCI-CT")[0]
    ad_sample = np.random.choice(ad_idx, size=len(ad_idx), replace=True)
    ct_sample = np.random.choice(ct_idx, size=len(ct_idx), replace=True)
    indices = np.concatenate([ad_sample, ct_sample])
    np.random.shuffle(indices)
    return X.iloc[indices], y[indices]

# ======================
# MAIN LOOP
# ======================
for seed in range(n_seeds):
    print("\n" + "="*50)
    print(f"RUNNING SEED {seed}")
    print("="*50)

    cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=seed)
    le = LabelEncoder()
    le.fit(y_train)

    threshold_scores = []
    
    # --- CV THRESHOLD OPTIMIZATION (Su Lisbona) ---
    for t in thresholds:
        fold_scores = []
        for f_idx, (train_idx, val_idx) in enumerate(cv.split(X_train, y_train)):
            X_sub, y_sub = X_train.iloc[train_idx], y_train[train_idx]
            X_val, y_val = X_train.iloc[val_idx], y_train[val_idx]

            vip_matrix = np.zeros((n_iterations, X_sub.shape[1]))
            for i in range(n_iterations):
                X_boot, y_boot = stratified_bootstrap(X_sub, y_sub)
                pls = PLSRegression(n_components=2)
                pls.fit(X_boot, le.transform(y_boot))
                vip_scores = calculate_vip(pls, X_boot)
                vip_matrix[i] = (vip_scores > 1).astype(int)

            vip_freq = vip_matrix.mean(axis=0)
            sel_feats = X_sub.columns[vip_freq >= t]
            if len(sel_feats) == 0: sel_feats = [X_sub.columns[np.argmax(vip_freq)]]

            pls_cv = PLSRegression(n_components=2)
            pls_cv.fit(X_sub[sel_feats], le.transform(y_sub))
            y_pred_val = pls_cv.predict(X_val[sel_feats]).ravel()
            fold_scores.append(matthews_corrcoef(le.transform(y_val), (y_pred_val > 0.5).astype(int)))

        threshold_scores.append(np.mean(fold_scores))

    best_v_thr = thresholds[np.argmax(threshold_scores)]
    best_stability_thresholds.append(best_v_thr)

    # --- FINAL SELECTION (Lisbona Full) ---
    final_vip_matrix = np.zeros((n_iterationsf, X_train.shape[1]))
    for i in range(n_iterationsf):
        X_boot, y_boot = stratified_bootstrap(X_train, y_train)
        pls = PLSRegression(n_components=2)
        pls.fit(X_boot, le.transform(y_boot))
        final_vip_matrix[i] = (calculate_vip(pls, X_boot) > 1).astype(int)

    final_vip_freq = final_vip_matrix.mean(axis=0)
    top_features = X_train.columns[final_vip_freq >= best_v_thr]
    if len(top_features) == 0: top_features = [X_train.columns[np.argmax(final_vip_freq)]]
    
    # Aggiorna conteggi globali
    global_stable_counts_coimbraonlisbon.loc[top_features] += 1
    n_features_per_seed.append(len(top_features))

    # --- TEST EVALUATION (Su Coimbra) ---
    pls_final = PLSRegression(n_components=2)
    pls_final.fit(X_train[top_features], le.transform(y_train))
    y_pred_cont_test = pls_final.predict(X_test[top_features]).ravel()
    
    test_auc = roc_auc_score(le.transform(y_test), y_pred_cont_test)
    
    best_mcc_seed = -1.0
    best_d_thr_seed = None
    for d_thr in decision_thresholds:
        y_class_test = (y_pred_cont_test > d_thr).astype(int)
        mcc_val = matthews_corrcoef(le.transform(y_test), y_class_test)
        if mcc_val > best_mcc_seed:
            best_mcc_seed = mcc_val
            best_d_thr_seed = d_thr
    
    all_test_mcc.append(best_mcc_seed)
    all_test_auc.append(test_auc)
    best_decision_thresholds.append(best_d_thr_seed)

    # OUTPUT PER SEED
    print(f"RESULTS SEED {seed}:")
    print(f" - MCC: {best_mcc_seed:.3f} | AUC: {test_auc:.3f}")
    print(f" - Features: {len(top_features)}")
    print(f" - Best Stability Thr (VIP): {best_v_thr}")
    print(f" - Best Decision Thr: {best_d_thr_seed}")

# ======================
# FINAL PERFORMANCE SUMMARY
# ======================
print("\n" + "="*50)
print("FINAL CONSOLIDATED RESULTS")
print("="*50)
print(f"Mean MCC: {np.mean(all_test_mcc):.4f} (± {np.std(all_test_mcc):.4f})")
print(f"Mean AUC: {np.mean(all_test_auc):.4f} (± {np.std(all_test_auc):.4f})")
print(f"Mean Features: {np.mean(n_features_per_seed):.2f} (± {np.std(n_features_per_seed):.2f})")

# Frequenze Soglie
print("\nStability Threshold (VIP) Frequency:")
print(pd.Series(best_stability_thresholds).value_counts(normalize=True).sort_index())

print("\nDecision Threshold Frequency:")
print(pd.Series(best_decision_thresholds).value_counts(normalize=True).sort_index())

df_plsda_CoimbraonLisbon = pd.DataFrame({
    "model": "PLS-DA",
    "seed": list(range(n_seeds)),
    "mcc": all_test_mcc,
    "auc": all_test_auc,
    "n_features": all_n_features,
    "best_vip_threshold": all_best_vip_thr,
    "best_decision_threshold": all_best_dec_thr
})

# TOP 30 PROTEINE (con nuovo nome variabile)
print("\n--- TOP 30 MOST STABLE PROTEINS (global_stable_counts_coimbraonlisbon) ---")
top_30_stability_CoimbraonLisbon = (global_stable_counts_coimbraonlisbon / n_seeds).sort_values(ascending=False)
```
